```python
def visualize(filename):
    # Load Data from txt files
    dataX = []
    dataY = []
    print(filename)
    with open(filename, 'r') as f:
        input = f.read()
        output = []
        for i in input.splitlines():
            dataX.append(i.split(', ')[0])
            dataY.append(i.split(', ')[1])

    # Visualize Data
    import math
    import matplotlib.pyplot as plt
    import numpy as np

    ax = plt.subplot(projection='polar')
    ax.grid(False)

    # print(lidarData)
    dataX = [float(d)/180 * math.pi for d in dataX]
    dataY = [float(d )for d in dataY]
    ax.plot(dataX, dataY, "b.", markersize = 0.5)
    ax.set_theta_direction(-1)
    ax.set_rmax(1800)
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.set_theta_offset(np.pi / 2.0)
    plt.savefig(filename.replace(".txt", ".png"))
    plt.show()
visualize("./0.txt")
visualize("./1.txt")
visualize("./2.txt")
visualize("./3.txt")
```
